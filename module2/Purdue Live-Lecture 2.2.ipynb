```python
# https://matplotlib.org/stable/api/axes_api.html

#read the file
agland_WORLD = pd.read_csv('data/faostat-aglanduse-world.csv')

#size the entire figure
fig1, ax = plt.subplots(figsize=(10,15))

#create the plot
plt.plot(agland_WORLD['Year'], agland_WORLD['Value']/1000, marker=".")

# #plot meta
ax.ticklabel_format(style='plain')
ax.set_box_aspect(0.4)
ax.set_ylim(4400, 5000)

plt.ylabel('1000 ha')
plt.grid(axis='y')
plt.title('Global Agricultural land - Area (1000 ha) from 1961-2019')

plt.show()
```
