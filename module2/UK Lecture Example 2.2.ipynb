```python
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
#read the file
agland_WORLD = pd.read_csv('FAOSTAT_data_9-14-2021.csv')

#create figure
fig1, ax = plt.subplots(figsize=(10,15))

#plot the data

ax.plot(agland_WORLD['Year'], agland_WORLD['Value']/1000, marker=".")

# display characteristics
ax.ticklabel_format(style='plain')
ax.set_box_aspect(0.4)
ax.set_ylim(4400, 5000)

ax.set_ylabel('1000 ha')
ax.grid(axis='y')
ax.set_title('Global Agricultural Land - Area (1000 ha) from 1961 to 2019')
ax.xaxis.set_major_locator(MultipleLocator(2))
plt.xticks(rotation=45)
plt.show()
```
